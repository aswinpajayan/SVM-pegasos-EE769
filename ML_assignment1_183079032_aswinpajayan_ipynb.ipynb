#SVM Classifier
Train a SVM classifier using gradient descent and return a Weight Matrix which is a numpy array of length (N + 1) where N is dimension of training samples. 
You can refer to Fig. 1 in [this](https://www.cs.huji.ac.il/~shais/papers/ShalevSiSrCo10.pdf) paper for implementation. You can add arguments to svm_function according to your implementation.

Examined two approaches for including the bias term. First approach adds
bias term in the w vector itslef by adding an extra dimension 
$\phi_{0}(x_i)=1$. But be should not be include in the weight regularisaton. One way of doing this is by setting $\phi_{0}x_i = 10$, i.e. a higher constant value such that the bias wont have much effect in the weight regularistaion. The second method , which is used here is to compare bias term seperately and prepare a seperate update step for the bias term using the standard gradient descent eq: $b_{t+1} = b_t - \eta_t * \nabla_t$ where $\eta_t = 
\frac{1}{\lambda t}$ . Bias update step was found out to be 
$$b_{t+1} = b_t(1 - \frac{1}{t}) + \eta_t * y_i ;\space  w^{T}\phi(x)+b <1$$ ,$$b_{t+1} = b_t(1 - \frac{1}{t});\space otherwise$$




```python
#this function includes bias inside w
#by increasing an extra dimesnsion in x

def svm_function(x, y, epoch, l_rate):
    w = np.array([0,0,0])
    for t in np.arange(1,epoch):
      eta_t = 1/(t*l_rate)
      # randomly choose a data point 
      i = np.random.randint(0,len(x))

      if (y[i] * (np.dot(x[i],w) ) < 1):
        # non zero loss case 
        w = (1 - (1/t)) * w + eta_t * y[i] * x[i]
      elif (y[i] * np.dot(x[i],w) >= 1):
         w = (1 - 1/float(t))*w
    return w

def svm_fx_bias(x, y, epoch, l_rate):
    b = 0
    w = np.array([0,0])
    for t in np.arange(1,epoch):
      
      eta_t = 1/(t*l_rate)
      # randomly choose a data point 
      i = np.random.randint(0,len(x))

      if (y[i] * (np.dot(x[i],w) + b ) < 1):
        # non zero loss case 
        w = (1 - (1/t)) * w + eta_t * y[i] * x[i]
        b = (1 -1/t)* b + eta_t*y[i]
      elif (y[i] *( np.dot(x[i],w) + b)>= 1):
        # zero loss
        w = (1 - 1/float(t))*w
        b = (1 -1/t)* b 
        
    w = np.hstack((w,b))
    return w

```

#Train a SVM classifier on the linearly non-separable data by appropriate features crafted from input data
For linearly non-separable data, you need to transform the data in a space where it can be linearly seprable. These features can be exponential, polynomial, trignometric or any other function of actual input features. For example, if your input data is (x1, x2) you can have hand-crafted features as (sin(x1), cos(x1), cos(x2), x1-x2). 
Here you need to think of which hand-crafted features can be best suited for data given to you. Write a function to convert input features to hand-crafted features. Use these features to train a SVM using svm_function.
Note that, if you choose to have L hand-crafted features, SVM will return L+1 dimensional 'w'. 


Looking at the scatter plot of labelled data , It seems like the classes can be seperated by some parabola of the form $x^2 + 4ay +b =0$ Features where thus chosen to be $$\phi_0(x_i) = x_{i1}^2$$
$$\phi_1(x_i) = x_{i2}$$

```python

def transform(data):
  x,y = data[:,:2],data[:,2]
  phi = np.column_stack((data,np.power(x[:,0],2),np.power(x[:,1],1)))
  return phi

phi = transform(data)
clr  = ['red' if(label == -1) else 'blue' for label in phi[:,2]]
plt.scatter(phi[:,3],phi[:,4],color = clr)
plt.show()
w = svm_fx_bias(phi[:,3:5],y,50000,0.001)
```
